```python
xls=[]
vcf=[]
txt=[]
for _ in work :
    if _.endswith('.xls'):
        print (len(_) - 4, _)
        xls.append(_)
    elif _.endswith('.txt'):
        print (len(_) -4 , _)
        txt.append(_)
    elif _.endswith('.vcf'):
        print (len(_)-4, _)
        vcf.append(_)
    else :
        print ('')
print(xls, len(xls))
print(vcf,len(vcf))
print(txt,len(txt))
```
